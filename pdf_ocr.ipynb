```python
import os
from doctr.io import DocumentFile
from doctr.models import ocr_predictor
```

```python
def process_pdfs_with_doctr(input_folder, output_folder):
  """
  Processes PDF files in the input folder using Doctr and saves the extracted text
  to the output folder.

  Args:
    input_folder: Path to the folder containing PDF files.
    output_folder: Path to the folder where output text files will be saved.
  """

  # Create the output folder if it doesn't exist
  os.makedirs(output_folder, exist_ok=True)

  # Load the most powerful pre-trained OCR model from Doctr
  model = ocr_predictor(pretrained=True)

  for filename in os.listdir(input_folder):
    if filename.endswith(".pdf"):
      pdf_path = os.path.join(input_folder, filename)

      try:
        # Load the PDF document
        doc = DocumentFile.from_pdf(pdf_path)

        # Perform OCR
        result = model(doc)

        # Extract the text
        text = result.render()

        # Save the extracted text to a file
        output_filename = os.path.splitext(filename)[0] + ".txt"
        output_path = os.path.join(output_folder, output_filename)
        with open(output_path, "w") as f:
          f.write(text)

        print(f"Processed {filename} and saved to {output_path}")

      except Exception as e:
        print(f"Error processing {filename}: {e}")

# Set the input and output folder paths
input_folder = 'soy_pdfs'
output_folder = 'output_text'

# Process the PDFs
process_pdfs_with_doctr(input_folder, output_folder)
```
