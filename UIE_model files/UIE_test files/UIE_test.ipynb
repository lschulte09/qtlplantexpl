```python
import pandas as pd
df = pd.read_csv("uie_manual_eval/entity_fn.csv")

tp_count = 0
fn_count = 0

for index, row in df.iterrows():
    # Extract columns of interest from the row
    text = row.get("text", "")
    gold_entity_type = row.get("gold_entity_type", "")
    gold_entity_text = row.get("gold_entity_text", "")
    all_pred_entities = row.get("all_pred_entities", "")

    # Display
    print("\n---------------------------------------------------------")
    print(f"Row index: {index}")
    print("Text: ", text)
    print("Gold Entity Type: ", gold_entity_type)
    print("Gold Entity Text: ", gold_entity_text)
    print("All Predicted Entities: ", all_pred_entities)

    # Ask user input: 'y' or 'n'
    choice = input(
        "Is this gold entity sufficiently covered by one of the predicted entities? (y/n) "
    ).strip().lower()

    if choice == 'y':
        tp_count += 1
    else:
        fn_count += 1

# Final summary
print("\n=== Manual Review Complete ===")
print(f"TP Count (covered)   = {tp_count}")
print(f"FN Count (not covered) = {fn_count}")
```

```python
"""
Reads 'entity_fp.csv' and for each row:
  - Prints out relevant columns.
  - Asks user if the predicted entity is valid.
      * 'y' => increment tp_extra_count
      * 'n' => increment fp_count
      * 'c' => increment covered_count
Finally, prints the totals.
"""
df = pd.read_csv("uie_manual_eval/entity_fp.csv")

tp_extra_count = 0  # how many FPs were actually correct (missing from gold)
fp_count = 0        # how many FPs are truly invalid
covered_count = 0   # how many FPs are partially covered by gold in some other way

for index, row in df.iterrows():
    # Extract columns of interest from the row
    text = row.get("text", "")
    pred_entity_type = row.get("pred_entity_type", "")
    pred_entity_text = row.get("pred_entity_text", "")
    all_gold_entities = row.get("all_gold_entities", "")

    # Display
    print("\n---------------------------------------------------------")
    print(f"Row index: {index}")
    print("Text: ", text)
    print("Predicted Entity Type: ", pred_entity_type)
    print("Predicted Entity Text: ", pred_entity_text)
    print("All Gold Entities: ", all_gold_entities)

    # Prompt user
    choice = input(
        "[y] if the prediction is valid and should have been in gold\n"
        "[n] if the prediction is truly wrong\n"
        "[c] if it is covered by an existing gold entity already\n"
        "=> Your choice (y/n/c)? "
    ).strip().lower()

    if choice == 'y':
        tp_extra_count += 1
    elif choice == 'c':
        covered_count += 1
    else:  # default to 'n'
        fp_count += 1

# Final summary
print("\n=== Manual Review Complete ===")
print(f"TP-extra Count (valid but missing in gold) = {tp_extra_count}")
print(f"Covered Count                              = {covered_count}")
print(f"FP Count (truly invalid)                   = {fp_count}")
```
